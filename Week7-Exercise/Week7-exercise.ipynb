- Find total number of movies playing on that day

```python
import bs4
import requests


r = requests.get('http://www.kultunaut.dk/perl/searchlist/type-nynaut?Genre=Familiefilm&Genre=Komedie%2FRomantik&Genre=Action%2FDrama&Genre=Science%20Fiction&Genre=Thriller%2FGyser&Genre=Tegnefilm%2FAnimation&Genre=Dokumentar&Genre=Andre%20Film&&Area=&periode=')
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')
movies_list = []

for movies in soup.find_all('b'):
    movies_list.append(movies.getText())

print("The total number of movies this day is: " + str(len(movies_list[1:])))
 
```

### Exercise 2 - webscraping
    
- Print all movie titles with their description

```python
moviesTable = soup.findAll('div', attrs={'style': 'clear:both'})

for movies in moviesTable:
    print(movies.getText())
    
```

- Find the number of all link on the site

```python
all_links = []

for link in soup.find_all('a'):
    if link.get('href')[0:4] == 'http':
        all_links.append(link)

print(len(all_links))
```
